```
import pandas as pd
```

##Movie Ratings

```
data = pd.read_csv("movie_ratings.csv")
data.head()
```

Quiero averiguar cuantos datos faltantes hay en nuestro dataframe

```
data.isnull().sum()
```

```
del data['Unnamed: 0']
```

```
data = data.drop_duplicates()
data
```

```
data['metascore']=data['metascore'].fillna(0.0)
data
```

```
data['imdb'].describe()
```

```
bins = (4.1,6.7,7.3,7.8,9.9)
group_names = ['Muy Mala','Mala','Buena','Excelente']
categories = pd.cut(data['imdb'], bins, labels=group_names)
data['Clasificacion'] = categories
data
```

```
data['Clasificacion'].value_counts()
```
